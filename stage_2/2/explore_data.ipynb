```python
import pandas as pd
import numpy as np
```

# Загружаем данные

```python
def split_data_to_X_y(df, y_cols):
    X = df.drop(y_cols, axis=1)
    y = df[y_cols]
    return X, y
```

## train

```python
df_train = pd.read_csv('data/Nti2018_task2Train.csv', index_col='index')

# удаляем ненужные колонки
cols_to_drop = ['brs_id']
df_train = df_train.drop(cols_to_drop, axis=1)

# находим колонки которые нужно предсказывать
y_cols = df_train.columns[df_train.columns.str.contains('Avg')]

print(df_train.shape)
df_train.head()
```

```python
X_train, y_train = split_data_to_X_y(df_train, y_cols)
print(X_train.shape, y_train.shape)
```

## test

```python
X_test = pd.read_csv('data/xTestFinal_2.csv', index_col='index')
print(X_test.shape)
X_test.head()
```
